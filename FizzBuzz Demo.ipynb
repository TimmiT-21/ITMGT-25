```python
for i in range (1,101):
    
    if ((i % 3) == 0) and ((i % 5) == 0):
        print('fizzbuzz')
    elif (i % 3) == 0:
        print('fizz')
    elif (i % 5) == 0:
        print('buzz')
    else: #added this later on
        print(i)
  
    ## My version, Sir suggests to put the more constrictive condition up front in the program. That way, it appears to be of concnetric circles of exclusivity, such that python checks what fulfills the smaller circles' coniditons first before the bigger, less restrictive cicrles. 
    ## Lessons learned
        1. #Iterative Programming, building blocks of code to make it more digestible and mutable
        2. #You won't get everything at the get-go. Just run your program, and take the lesson; matters more is debugging and fixing the error of the code.
```
